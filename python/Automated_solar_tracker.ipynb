```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
time = np.arange(6, 18.5, 0.5)

sun_angle = (time - 6) * 15

angle_rad = np.deg2rad(sun_angle)

sun_height = np.sin(angle_rad)
```

```python
# Plot the graph
plt.figure(figsize=(8,5))
plt.plot(time, sun_angle, marker='o', linewidth=2)

plt.title("Sun Angle vs Time")
plt.xlabel("Time (Hours)")
plt.ylabel("Sun Angle")
plt.xticks(np.arange(6, 19, 1))
plt.yticks(sun_angle)
plt.grid(True)

plt.show()
```

```python
# Plot the graph
plt.figure(figsize=(8,5))
plt.plot(time, sun_height, marker='o', linewidth=2)

plt.title("Sun Height vs Time")
plt.xlabel("Time (Hours)")
plt.ylabel("Relative Sun Height")
plt.xticks(np.arange(6, 19, 1))
plt.grid(True)

plt.show()
```
